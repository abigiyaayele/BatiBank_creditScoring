Distribution of Numerical Features

```python
def plot_numerical_distribution(data):
    numerical_features = data.select_dtypes(include=[np.number]).columns.tolist()
    print(f"Numerical Features: {numerical_features}")

    plt.figure(figsize=(20, 15))
    for i, feature in enumerate(numerical_features):
        plt.subplot(5, 4, i + 1)
        sns.histplot(data[feature], kde=True, bins=30)
        plt.title(f'Distribution of {feature}')
    plt.tight_layout()
    plt.show()

# Plot numerical feature distribution
plot_numerical_distribution(data)
```

Distribution of Categorical Features

```python
def plot_categorical_distribution(data):
    categorical_features = data.select_dtypes(include=['object']).columns.tolist()
    print(f"Categorical Features: {categorical_features}")

    for feature in categorical_features:
        plt.figure(figsize=(8, 6))
        sns.countplot(y=data[feature], order=data[feature].value_counts().index)
        plt.title(f'Distribution of {feature}')
        plt.xlabel('Count')
        plt.ylabel(feature)
        plt.show()

# Plot categorical feature distribution
plot_categorical_distribution(data)
```

Correlation Analysis

```python
def correlation_analysis(data):
    # Select only numeric columns
    numeric_data = data.select_dtypes(include=np.number)
    
    # Compute correlation matrix
    correlation_matrix = numeric_data.corr()
    
    # Plot heatmap
    plt.figure(figsize=(15, 10))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
    plt.title('Correlation Matrix')
    plt.show()

# Perform correlation analysis
correlation_analysis(data)
```

```python
def get_frequent_items(data):
    frequent_items = {}
    for column in data.columns:
        if column == 'PricingStrategy' or column == 'FraudResult':
            frequent_items[column] = data[column].mode().values[0]  # Calculate mode for specific columns
        elif data[column].dtype == 'float64' or data[column].dtype == 'int64':
            frequent_items[column] = data[column].mean()  # Calculate mean for numeric columns
        else:
            frequent_items[column] = data[column].mode().values[0]  # Calculate mode for non-numeric columns
    return frequent_items
get_frequent_items(data)
```

```python
# Assuming 'data' is the DataFrame containing the data

frequent_items = get_frequent_items(data)
```

```python
def plot_counts_per_column(labels, counts):
    
    plt.bar(labels, counts)
    plt.xticks(rotation='vertical')
    plt.xlabel('Columns')
    plt.ylabel('Counts')
    plt.title('Counts per Column')
    plt.show()
```

```python
# Plot counts per column
labels = ['TransactionId', 'BatchId', 'AccountId', 'SubscriptionId', 'CustomerId', 'CurrencyCode', 'CountryCode',
          'ProviderId', 'ProductId', 'ProductCategory', 'ChannelId', 'TransactionStartTime', 'PricingStrategy',
          'FraudResult']
counts = [1, 28, 30893, 32630, 4091, 95662, 95662, 38189, 32635, 45405, 56935, 17, 79848, 95469]
plot_counts_per_column(labels, counts)
```

Identify missing values

```python
def identify_missing_values(data):
    missing_values = data.isnull().sum()
    print("Missing Values:")
    print(missing_values[missing_values > 0])
    print("\n")
# Identify missing values
identify_missing_values(data)
```

Outlier Detection

```python
def detect_outliers(data):
    numerical_features = data.select_dtypes(include=[np.number]).columns.tolist()
    print(f"Numerical Features: {numerical_features}")
    
    for feature in numerical_features:
        plt.figure(figsize=(8, 6))
        sns.boxplot(data[feature])
        plt.title(f'Box plot of {feature}')
        plt.show()

# Detect outliers
detect_outliers(data)
```
